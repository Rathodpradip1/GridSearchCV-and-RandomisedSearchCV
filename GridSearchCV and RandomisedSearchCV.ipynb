```python
# importing the dependancies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
# loading dataset from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# Loading the data to pandas dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data , columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# Adding target column to the data frame
data_frame["label"] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
# checking the number of rows and columns in the dataset
data_frame.shape
```

```python
# checking missing value
data_frame.isnull().sum()
```

```python
data_frame.describe()
```

```python
data_frame.groupby("label").mean()
```

```python
# checking distribution of target variable
data_frame["label"].value_counts()
```

```python
# seoerating feature and target
X = data_frame.drop("label" , axis = 1)
Y = data_frame["label"]
```

```python
print(X)
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

```python
# GridSearchCV --> gridsearchcv is used to determine the best parameter for the modal
```

```python
# loading the SVC model
model = SVC()
```

```python
# hyperparameter

parameters = { 
               'kernel':['linear','poly','rbf', 'sigmoid'],
                'C':[1, 5, 10,20]
}
```

```python
# gridsearch
classifier = GridSearchCV(model , parameters, cv=5)
```

```python
# fitting data to model
classifier.fit(X , Y)
```

```python
classifier.cv_results_
```

```python
# Best_parameters

best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# highest_accuracy
highest_accuracy = classifier.best_score_

print(highest_accuracy)
```

```python
result = pd.DataFrame(classifier.cv_results_)

result.head()
```

```python
grid_search_result = result[['param_C' , 'param_kernel' , 'mean_test_score']]
```

```python
grid_search_result
```

```python
# Highyest_accuracy = 95.2%
# best_parameters = (C:10 , 'kernel':'linear')
```

```python
# RandomisedSearchCV
```

```python
#Loading SVC model
model = SVC()
```

```python
# hyperparameters

parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
# Randomised search
classifier = RandomizedSearchCV(model , parameters , cv=5)
```

```python
# fitting data into the model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# best_parameter
classifier.best_params_
```

```python
# highyest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# loading the result into pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
randomised_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
print(randomised_search_result)
```
